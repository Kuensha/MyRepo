````
edad = 35

if edad<18:
    print("Sos menor de edad")    
elif edad<25:
    print("Sos un adulto joven")
elif edad<40:
    print("Estas en la flor de la vida!!!")
    print("Esta linea es para fines academicos")
elif edad<55:
    print("Sos muuy joven (si no Gustavo me mata!)")
else:
    print("Ya vas llegando a la jubiliacion")
````

**CICLOS WHILE**

---

**Spoiler**: Warning para los puristas de la programación!! vamos a ver maneras de salir anticipadamente de un ciclo.

````
myCont = 15

# Cuidado!! si no actualizamos la variable de control, podemos caer en un ciclo infinito
while myCont<25:
    print(f"iteracion {myCont-14} : {myCont}")
    myCont +=1
````

````
# ROMPER UN CICLO ANTICIPADAMENTE
# LOS INGENIEROS DE SOFTWARE O PICHONES DE PROGRAMADORES ME ODIARÁN...MUAHAHAHAHAHA
# PODEMOS SALIR ANTICIPADAMENTE DE UN CICLO MEDIANTE LA SENTENCIA break

myCont2 = 0

# Cuidado!! si no actualizamos la variable de control, podemos caer en un ciclo infinito
while myCont2<100:  # esto contaria 100 iteraciones
    print(f"iteracion {myCont2}")
    myCont2 +=1
    if myCont2==10: # pero cuando llegue a 10, saldrá anticipadamente
        # el 10 no lo va a mostrar
        break
````

````
# PODEMOS NO PROCESAR DETERMINADAS ITERACIONES CON LA SENTENCIA continue
# QUIERO MOSTRAR TODOS LOS NUMEROS DEL 0 AL 20, PERO LE TENGO FOBIA A LOS MULTIPLOS DE 3
# ASI QUE NO LOS QUIERO MOSTRAR
myCont3 = -1
while myCont3<21:
    myCont3+=1
    if (myCont3%3==0):
        continue
    print(myCont3)
    
````

**CICLOS FOR**

---
LOS CICLOS FOR SE UTILIZAN PRINCIPALMENTE PARA ITERAR SOBRE SECUENCIAS (LISTAS, TUPLAS, DICCIONARIOS, SETS).

NO SE "PARECE TANTO"  A LOS CICLOS **FOR** DE LA MAYORÍA DE LOS LENGUAJES DE PROGRAMACIÓN SINO MAS BIEN A ALGUNOS MÉTODOS ITERADORES ENCONTRADOS EN ALGUNOS LENGUAJES ORIENTADOS A OBJETOS. 

---


````
# ITERAR SOBRE LOS ELEMENTOS DE UNA LISTA
myString = "Esto es una oracion que voy a utilizar con fines demostrativos"
myNewString = 'Esto-tambien-es-una-oracion-de-practica'

# ANTES DE ARRANCAR CON LOS CICLOS FOR VAMOS A VER UN METDODO MUY UTIL
# EL METODO SPLIT GENERA UNA LISTA A PARTIR DE UN STRING
# POR DEFECTO UTILIZA COMO DELIMITADOR LOS ESPACIOS EN BLANCO, PERO ES POSIBLE
# INDICARLE UN NUEVO DELIMITADOR PONIENDOLO ENTRE PARENTESIS
#
miLista = myString.split() 
miLista2 = myNewString.split("-") 

print("miLista = ", miLista)
print("miLista2 = ", miLista2)
````

````
# AHORA SI, VAMOS A VER COMO MOSTRAR LOS ELEMENTOS DE LAS LISTAS, DE A 1 POR VEZ

#  CICLO FOR ESTILO C O MATLAB
# PARA ESTO DEBEMOS UTILIZAR "RANGE" Y "LEN"
# RANGE YA LO VIMOS Y LEN DEVUELVE LA LONGITUD DE LA LISTA
for i in range(len(miLista)):
    print(miLista[i])
````

````
#PERO LOS CICLOS FOR SE PUEDEN USAR DE UNA MANERA MUCHO MEJOR!!!

for palabras in miLista2:
    print(palabras)
# Y NO ES NECESARIO USAR RANGE Y LEN
````

````
# TAMBIEN PODEMOS USAR FOR PARA RECORRER DICCIONARIOS
myDict500 = {"asignatura":"Inteligencia Computacional", "carrera":"Computacion", "horas":5}

#HAY VARIAS MANERAS
# MANERA 1 - UTILIZANDO DICTIONARY.ITEMS()
for key, value in myDict500.items():
    print(f"{key}:  {value} ")

print("\n\n")
# MANERA 2 
for key in myDict500:
    print(f"{key}:  {myDict500[key]} ")

print("\n\n")
# MANERA 3
for key in myDict500.keys():
    print(f"{key}:  {myDict500[key]} ")

# Y SEGURAMENTE HAY MAS MANERAS PERO SE ME OCURRIERON ESTAS
````
