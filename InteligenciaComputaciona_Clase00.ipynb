**NÚMEROS FLOTANTES**

---

````
# VEAMOS CON LOS FLOTANTES QUE OCURRE
myFloat = 10.5
print("El tipo de dato de la variable myFloat es: ", type(myFloat))

# AHHHH, PERO HABIAMOS DICHO QUE PYTHON TENIA TIPADO DINAMICO NO?
# ASIGNEMOS UN FLOTANTE A myInt PARA VER QUE OCURRE
myInt = 18.5
print("El tipo de dato de la variable myInt es: ", type(myInt))
# QUE PASO??? AHORA myInt YA NO TIENE UN ENTERO ALMACENADO
````

````
# CUIDADO CON LAS OPERACIONES ENTRE DISTINTOS TIPOS...PUEDEN OCURRIR COSAS RARAS
print("Si sumo un entero mas un flotante, el resultado es...", type(5 + 5.0))
# correcto, flotante!

print("Si multiplico un entero y un flotante, el resultado es...", type(5 * 5.0))
# y si, flotante!

print("Si divido dos enteros obviamente me va a dar...", type(5/2))
# obvio...flotante!

# WAIT! COMO HAGO PARA DIVIDIR DOS ENTEROS Y SOLO OBTENER LA PARTE ENTERA
print("Si divido dos enteros utilizando doble slash // el resultado es...", type(5//2))
# ENTERO!!
````

**STRINGS**

---
***Spoiler 1:*** En Python se indexa desde cero<br>
***Spoiler 2:*** En Python los strings son inmutables, esto quiere decir que no podemos modificar mediante asignación ninguno de sus componentes

---

````
# STRINGS
myString1 = "Esto es un String"
myString2 = 'Esto tambien es un String'

print("El tipo de dato de la variable myString1 es: ", type(myString1))
print("El tipo de dato de la variable myString2 es: ", type(myString2))
````

````
# PARA QUE SIRVE QUE PODAMOS USAR COMILLAS DOBLES O SIMPLES ?
myString3 = "Uso comillas dobles para poder usar comillas simples ' dentro "
myString4 = 'Uso comillas simples para poder usar comillas dobles " dentro '

print("\n"+myString3)
print("\n"+myString4)

# ALTOOOO NOELIA!!! QUE ES ESO DE "\n" + myString3
# EL SIGNO + SIRVE PARA CONCATENAR CADENAS
# \n ES UNA SECUENCIA DE ESCAPE QUE SIMBOLIZA UNA "NUEVA LINEA"
````

````
# PUEDO ACCEDER A UNA POSICION DETERMINADA DEL STRING ? 
print("Vamos a acceder a la primera posicion : ", myString1[0])
# UN MOMENTO!!!! DIJISTE LA PRIMERA POSICION PERO USASTE EL INDICE CERO ??
# EXACTO! EN PYTHON SE INDEXA DESDE EL CERO Y NO DESDE EL UNO

# MMMMM....SI PUEDO VER EL CONTENIDO DE UNA POSICION USANDO CORCHETES...
# ¿PUEDO MODIFICAR UN VALOR HACIENDO LO MISMO?
myString1[0] = 'P'

# AUCH!! UN ERROR...QUE HA OCURRIDO???
# VAMOS POR PARTES PARA SACARLE JUGO A ESTE ERROR
# 1) PUEDEN NOTAR QUE AUNQUE HAY UN ERROR EN EL SCRIPT...EL MISMO SE EJECUTA 
# SIN PROBLEMA HASTA LA LINEA 7 --> ESTO ES DEBIDO A QUE ES UN LENGUAJE INTERPRETADO
# 2) EL ERROR ES DEBIDO A QUE ESTAMOS INTENTANDO MODIFICAR MEDIANTE ASIGNACION
# UN STRING --> LOS STRINGS SON INMUTABLES!!! NO SOPORTAN QUE LOS CAMBIEN
````

**LISTAS**

---

````
# LISTAS
# LAS LISTAS PUEDEN ALMACENAR MUUUCHOS DATOS
myList1 = [1, 2, 3, 4, 5] 
# PERO TAMBIEN PUEDEN CONTENER DATOS MIXTOS
myList2 = [1, 2, "Noelia es la mejor", 5+4j, 'perro']
# Incluso, pueden tener listas dentro
myList3 = [1, 2, ["Noelia", "es","la", 'mejor']]

print("Lista 1: ",myList1)
print("\nLista 2: ",myList2)
print("\nLista 3: ",myList3)
````

````
# COMO ACCEDEMOS A UN ELEMENTO EN PARTICULAR DE UNA LISTA ?
print("Elemento 2 de la lista 1: ", myList1[1])
print("Elemento 3 de la lista 2: ", myList2[2])
````

````
# PUEDO MODIFICAR UN ELEMENTO DE UNA LISTA ??
myList2[4] = 'gato'
print("\nModifique la lista2: ", myList2)
# SIII, SE PUEDE!! LAS LISTAS SON MUTABLES!!!
````

````
# PUEDO AGREGAR UN DATO NUEVO? 
myList2[5] = "El dato nuevo"
# AUCH! NOELIA, DIJISTE QUE ERAN MUTABLES PERO ES MENTIRA!!
# DICE "FUERA DE RANGO"
````

````
# SE PUEDEN AGREGAR DATOS A UNA LISTA, PERO DEBEMOS HACERLO DE OTRA MANERA
# MEDIANTE EL METODO APPEND
myList2.append("El dato nuevo")
print("\nAgregue un dato nuevo mediante el metodo append: ", myList2)
````

````
# CON EL METODO APPEND PUEDO AGREGAR UNA LISTA AL FINAL DE OTRA ??? PUES NO
# DADO QUE ES UNA CONCATENACION, SE UTILIZA EL +
myList1 += myList3
print("\nMi pobre lista 1 ahora es enorme!!! ", myList1)
````
