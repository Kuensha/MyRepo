COMO PLOTEAR UN VECTOR DE DATOS


---


```
# GENERO UN VECTOR DE DATOS ENTRE 0 Y 9
t = range(10) # range(start, stop, step)
# GENERO UNA LISTA CON LOS VALORES AL CUADRADO DE t
y = [y**2 for y in t]   # --> ESTO SE LLAMA COMPREHENSION LIST

print(t)
print(y)
# QUE OCURRE PORQUE NO SE MUESTRAN LOS VALORES DE t ? 
# EN PYTHON 3, RANGE NO ES UNA BUILT-IN FUNCTION, ES UN TIPO DE DATOS
```

```
# COMO HACEMOS PARA VER LOS DATOS GENERADOS ? --> CASTEAMOS EL TIPO DE DATOS 
# A UNA LISTA
t_l = list(t)
print(t_l)
```

```
# COMO PODEMOS GRAFICAR ESTOS DATOS ??
# NECESITAMOS IMPORTAR UN MODULO DEDICADO
import matplotlib.pyplot as plt

plt.figure(1,figsize = (7,7))
plt.plot(t, y)
plt.show()
```

```
# AGREGAMOS UN TITULO
plt.figure(2,figsize = (7,7))
plt.plot(t, y)
plt.title("Nuestro primer plot")
plt.show()
```

```
plt.figure(3,figsize = (7,7))
plt.plot(t, y)
plt.title("Nuestro primer plot")
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([0, 10])
plt.ylim([0, 100])
plt.grid(True)
plt.show()
```
